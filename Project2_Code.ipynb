```python
import pandas as pd
import numpy as np
project_name = "Project 2: Automation of NDOT Concrete Mix Design"
client = "Nebraska Department of Transportation (NDOT)"
print(f"The project name is {project_name}")
print(f"The client is {client}")
```

```python
cubic_yard_ft3 = 27
unit_weight_water = 62.4
```

```python
def calculate_water_weight(cement_A, fly_ash_B, silica_fume_C, SCM_D, wc_ratio_E):
    total_cementitious = cement_A + fly_ash_B + silica_fume_C + SCM_D
    water_weight = total_cementitious * wc_ratio_E
    return water_weight
```

```python
#cement volume calculation
def calculate_vol_of_cement(cement_A, gravity_cement_J):
    gravity_water = gravity_cement_J * unit_weight_water
    vol_of_cement = cement_A / gravity_water
    return vol_of_cement
```

```python
#fly ash volume calculation
def calc_vol_fa(cement_A, gravity_cement_J):
    gravity_water = gravity_cement_J * unit_weight_water
    vol_of_fa = cement_A / gravity_water
    return vol_of_fa
```

```python
#silica fume volume calculation
def calculate_vol_T(silica_A, sg_silica):
    return silica_A / (sg_silica * unit_weight_water)
```

```python
#other SCM volume calculation
def calculate_vol_U(scm_A, sg_scm):
    return scm_A / (sg_scm * unit_weight_water)
```

```python
#air volume calculation
def calc_air_vol(air_content):
    air_vol = (air_content / 100) * cubic_yard_ft3
    return air_vol
```

```python
#water volume calculation
def calc_water_vol(weight_water):
    W = weight_water / unit_weight_water
    return W
```

```python
#total aggregate volume calculation
def calc_agg_vol(vol_R, vol_S, vol_T, vol_U, vol_V, vol_W):
    vol_X = cubic_yard_ft3 - vol_R - vol_S - vol_T - vol_U - vol_V - vol_W
    return vol_X
```

```python
#fine aggregate weight calculation
def calc_fine_agg(percent_fine_G, sg_fine_N, vol_X):
    weight_Y = unit_weight_water * (percent_fine_G / 100) * sg_fine_N * vol_X
    return weight_Y
```

```python
#coarse aggregate weight calculation
def calc_coarse_agg(percent_coarse_H, sg_coarse_0, vol_X):
    return unit_weight_water * (percent_coarse_H / 100) * sg_coarse_0 * vol_X
```

```python
#other aggregate weight calculation
def calc_other_agg(percent_other_I, sg_other_P, vol_X):
    return unit_weight_water * (percent_other_I / 100) * sg_other_P * vol_X
```

```python
# General info
project_no = int(input("Enter project number: "))
concrete_class = input("Enter class of concrete:")

# Cementitious material inputs
cement_A = float(input("Enter cement weight A (lb per cubic yard): "))
fly_ash_B = float(input("Enter fly ash weight B (lb cubic yard): "))
silica_fume_C = float(input("Enter silica fume weight C (lb cubic yard): "))
other_scm_D = float(input("Enter other SCM weight D (lb cubic yard): "))

# Design parameters
water_cement_ratio_E = float(input("Enter target water-cement ratio E:"))
air_content_F = float(input("Enter target air content F (%): "))

# Aggregate proportions
percent_fine_G = float(input("Enter percent fine aggregate G (%): "))
percent_coarse_H = float(input("Enter percent coarse aggregate H (%): "))
percent_other_I = float(input("Enter percent other SCM I (%): "))

# Specific gravities
sg_cement_J = float(input("Enter specific gravity of cement J: "))
sg_fly_ash_K = float(input("Enter specific gravity of fly ash K: "))
sg_silica_fume_L = float(input("Enter specific gravity of silica fume L: "))
sg_other_scm_M = float(input("Enter specific gravity of other SCM: "))

sg_fine_N = float(input("Enter specific gravity of fine aggregate N: "))
sg_coarse_O = float(input("Enter specific gravity of coarse aggregate O: "))
sg_other_P = float(input("Enter specific gravity of other aggregate P: "))

# A Print message to confirm all inputs have been collected
print("\nAll user inputs collected successfully.")
```

```python
Q = calculate_water_weight(cement_A, fly_ash_B, silica_fume_C, other_scm_D, water_cement_ratio_E)
```

```python
R = calculate_vol_of_cement(cement_A, sg_cement_J)
S = calc_vol_fa(fly_ash_B, sg_fly_ash_K)
T = calculate_vol_T(silica_fume_C, sg_silica_fume_L)
U = calculate_vol_U(other_scm_D, sg_other_scm_M)
```

```python
V = calc_air_vol(air_content_F)
W = calc_water_vol(Q)
```

```python
X = calc_agg_vol(R,S,T,U,V,W)
```

```python
Y = calc_fine_agg(percent_fine_G, sg_fine_N, X)
Z = calc_coarse_agg(percent_coarse_H, sg_coarse_O, X)
AA = calc_other_agg(percent_other_I, sg_other_P, X)
```
